# Datensatz 4 – Temperaturmessung auf dem Balkon über ein Jahr


## Beschreibung des Datensatzes (R4.1)

Die Temperaturwerte stellen angenommene Tagesmittelwerte der
Außentemperatur dar, wie sie auf einem Balkon gemessen werden könnten.
Der betrachtete Zeitraum umfasst ein Jahr mit täglichen Messpunkten.
Die Messwerte wurden im Rahmen dieser Projektarbeit selbst generiert und
orientieren sich an realistischen jahreszeitlichen Temperaturverläufen.



```python
import numpy as np
import pandas as pd

# Reproduzierbarkeit
np.random.seed(42)

# Anzahl der Tage
tage = 365
t = np.arange(tage)

# Jahreszeitlicher Temperaturverlauf (Sinus)
mittelwert = 12        # mittlere Jahrestemperatur in °C
amplitude = 10         # Jahresamplitude
rauschen = 2           # zufällige tägliche Schwankung

temperatur = (
    mittelwert
    + amplitude * np.sin(2 * np.pi * t / tage)
    + np.random.normal(0, rauschen, tage)
)

# DataFrame erstellen
df4 = pd.DataFrame({
    "tag": t + 1,
    "temperatur_celsius": temperatur
})

df4.head(), df4.shape
```

```python
df4.to_csv("../data/raw/datensatz_4_temperatur_balkon.csv", index=False)
```

## Lage- und Streuungskennwerte (R4.5–R4.6)

Der arithmetische Mittelwert und der Median der Temperaturwerte liegen
nah beieinander, was auf eine weitgehend symmetrische Verteilung der
Messwerte hindeutet. Der Modus ist aufgrund der stetigen Messwerte nicht
eindeutig bestimmt.

Die berechnete Stichprobenvarianz beschreibt die Streuung der
Temperaturwerte über den betrachteten Zeitraum und spiegelt die
jahreszeitlichen Schwankungen der Außentemperatur wider.


```python
# Lagekennwerte
modus = df4["temperatur_celsius"].mode()
mittelwert = df4["temperatur_celsius"].mean()
median = df4["temperatur_celsius"].median()

modus, mittelwert, median
```

```python
# Stichprobenvarianz
varianz = df4["temperatur_celsius"].var(ddof=1)
varianz
```

Die Auswertung der Lagekennwerte zeigt, dass der arithmetische
Mittelwert der Temperaturmessungen bei etwa 12,02 °C liegt. Der Median
beträgt etwa 12,34 °C und liegt damit nahe beim Mittelwert. Dies deutet
auf eine weitgehend symmetrische Verteilung der Temperaturwerte hin.

Ein eindeutiger Modus ist nicht vorhanden, da es sich um stetige
Messwerte handelt, bei denen einzelne Werte in der Regel nur einmal
auftreten.

Die Stichprobenvarianz von etwa 52,59 beschreibt die Streuung der
Temperaturwerte über den betrachteten Zeitraum. Die vergleichsweise
hohe Varianz ist durch den jahreszeitlichen Temperaturverlauf mit
niedrigen Winter- und hohen Sommertemperaturen erklärbar.
