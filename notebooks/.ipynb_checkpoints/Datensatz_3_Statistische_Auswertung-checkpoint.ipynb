## R3.3 – Datenbereinigung

In diesem Abschnitt werden die zusammengeführten Rohdaten bereinigt.
Dabei werden die Spaltennamen vereinheitlicht, die Zeichenkodierung
korrigiert sowie numerische Werte in geeignete Datentypen überführt.
Alle vorgenommenen Maßnahmen dienen der Vorbereitung der statistischen
Auswertung.


```python
import pandas as pd

# Bereinigte Arbeit beginnt ab der zusammengeführten Datei
pfad_raw = "../data/processed/datensatz_3_raw_merged.csv"

df = pd.read_csv(pfad_raw, sep=",", encoding="latin1")

df.head()
```

```python
# Originale Spaltennamen anzeigen
df.columns
```

```python
# Spaltennamen vereinheitlichen
df.columns = [
    "id",
    "wirtschaftszweig",
    "anzahl_ueberlebende",
    "ueberlebensrate_prozent"
]

df.head()
```

```python
df["wirtschaftszweig"] = df["wirtschaftszweig"].str.encode("latin1").str.decode("utf-8", errors="ignore")
```

```python
# Dezimalkomma durch Punkt ersetzen
df["anzahl_ueberlebende"] = (
    df["anzahl_ueberlebende"]
    .astype(str)
    .str.replace(",", ".", regex=False)
)

df["ueberlebensrate_prozent"] = (
    df["ueberlebensrate_prozent"]
    .astype(str)
    .str.replace(",", ".", regex=False)
)

# Umwandlung in numerische Datentypen
df["anzahl_ueberlebende"] = pd.to_numeric(df["anzahl_ueberlebende"], errors="coerce")
df["ueberlebensrate_prozent"] = pd.to_numeric(df["ueberlebensrate_prozent"], errors="coerce")

df.info()
```

```python
df.isna().sum()
```

```python
pfad_clean_csv = "../data/processed/datensatz_3_clean.csv"
df.to_csv(pfad_clean_csv, index=False)
pfad_clean_csv
```

```python
pfad_clean_xlsx = "../data/processed/datensatz_3.xlsx"
df.to_excel(pfad_clean_xlsx, index=False)
pfad_clean_xlsx
```

Im Rahmen der Datenbereinigung wurden die Spaltennamen vereinheitlicht
und technisch bedingte Sonderzeichen entfernt. Zudem wurde die
Zeichenkodierung korrigiert, sodass Umlaute korrekt dargestellt werden.

Die numerischen Variablen wurden aus Textformaten in geeignete numerische
Datentypen überführt. Dabei wurde das Dezimalkomma durch einen Dezimalpunkt
ersetzt. Fehlende oder nicht interpretierbare Werte wurden als solche
gekennzeichnet und nicht entfernt.

## R3.9 – Lagekennwerte

In diesem Abschnitt werden für jede numerische Variable die
Lagekennwerte Modus, arithmetischer Mittelwert und Median bestimmt.
Diese Kennwerte beschreiben die zentrale Lage der jeweiligen
Werteverteilung.


```python
# Lagekennwerte für Anzahl überlebender Unternehmen
modus_anzahl = df["anzahl_ueberlebende"].mode()
mittelwert_anzahl = df["anzahl_ueberlebende"].mean()
median_anzahl = df["anzahl_ueberlebende"].median()

# Lagekennwerte für Überlebensrate
modus_rate = df["ueberlebensrate_prozent"].mode()
mittelwert_rate = df["ueberlebensrate_prozent"].mean()
median_rate = df["ueberlebensrate_prozent"].median()

modus_anzahl, mittelwert_anzahl, median_anzahl, modus_rate, mittelwert_rate, median_rate
```

### Lagekennwerte der Variablen

- Arithmetischer Mittelwert: 65,04
- Median: 22,5

**Überlebensrate in Prozent**

- Median: 90,3


Der arithmetische Mittelwert beschreibt den durchschnittlichen Wert der
jeweiligen Variable. Der Median gibt den mittleren Wert der geordneten

## R3.10 – Spannweite

In diesem Abschnitt wird für jede numerische Variable die Spannweite
bestimmt. Die Spannweite beschreibt die maximale Streuung der Daten
und ergibt sich aus der Differenz zwischen dem größten und dem kleinsten
Wert.


```python
# Spannweite: Anzahl überlebender Unternehmen
spannweite_anzahl = df["anzahl_ueberlebende"].max() - df["anzahl_ueberlebende"].min()

# Spannweite: Überlebensrate in Prozent
spannweite_rate = df["ueberlebensrate_prozent"].max() - df["ueberlebensrate_prozent"].min()

spannweite_anzahl, spannweite_rate
```

### Spannweite der Variablen

**Anzahl überlebender Unternehmen**
- Spannweite: 313

**Überlebensrate in Prozent**
- Spannweite: 56,3


Die Spannweite gibt an, über welchen Wertebereich sich die jeweiligen
Messwerte erstrecken. Sie berücksichtigt ausschließlich die Extremwerte
und ist daher empfindlich gegenüber Ausreißern.


## R3.11 – Mittlere Abweichung vom Median

In diesem Abschnitt wird für jede numerische Variable die mittlere
absolute Abweichung vom Median bestimmt. Dieses Streuungsmaß beschreibt,
wie stark die einzelnen Werte im Durchschnitt vom Median abweichen.


```python
# Median berechnen
median_anzahl = df["anzahl_ueberlebende"].median()
median_rate = df["ueberlebensrate_prozent"].median()

# Mittlere absolute Abweichung vom Median
mad_anzahl = (df["anzahl_ueberlebende"] - median_anzahl).abs().mean()
mad_rate = (df["ueberlebensrate_prozent"] - median_rate).abs().mean()

mad_anzahl, mad_rate
```

### Mittlere Abweichung vom Median

**Anzahl überlebender Unternehmen**
- Mittlere Abweichung vom Median: 56,75

**Überlebensrate in Prozent**
- Mittlere Abweichung vom Median: 7,39


Die mittlere Abweichung vom Median beschreibt die durchschnittliche
Distanz der Messwerte zum Median. Im Vergleich zur Varianz oder
Standardabweichung ist dieses Streuungsmaß weniger anfällig für
Ausreißer und eignet sich besonders für schief verteilte Daten.


## R3.12 – Stichprobenvarianz

In diesem Abschnitt wird für jede numerische Variable die
Stichprobenvarianz bestimmt. Die Varianz ist ein Streuungsmaß und
beschreibt, wie stark die Messwerte um ihren Mittelwert streuen.


```python
# Stichprobenvarianz (ddof=1!)
varianz_anzahl = df["anzahl_ueberlebende"].var(ddof=1)
varianz_rate = df["ueberlebensrate_prozent"].var(ddof=1)

varianz_anzahl, varianz_rate
```

### Stichprobenvarianz der Variablen

**Anzahl überlebender Unternehmen**
- Stichprobenvarianz: 8023

**Überlebensrate in Prozent**
- Stichprobenvarianz: 120


Die Stichprobenvarianz beschreibt die mittlere quadratische Abweichung
der Messwerte vom arithmetischen Mittelwert. Durch die Division durch
(n−1) wird eine unverzerrte Schätzung der Varianz der Grundgesamtheit
ermöglicht.


## R3.13 – Variationskoeffizient

In diesem Abschnitt wird für jede numerische Variable der
Variationskoeffizient bestimmt. Dieser Kennwert beschreibt die relative
Streuung der Daten im Verhältnis zum arithmetischen Mittelwert.


```python
import numpy as np

# Standardabweichung aus Stichprobenvarianz
std_anzahl = np.sqrt(varianz_anzahl)
std_rate = np.sqrt(varianz_rate)

# Mittelwerte (bereits berechnet)
mean_anzahl = mittelwert_anzahl
mean_rate = mittelwert_rate

# Variationskoeffizient
vk_anzahl = std_anzahl / mean_anzahl
vk_rate = std_rate / mean_rate

vk_anzahl, vk_rate
```

### Variationskoeffizient der Variablen

**Anzahl überlebender Unternehmen**
- Variationskoeffizient: 1,38

**Überlebensrate in Prozent**
- Variationskoeffizient: 0,12


Der Variationskoeffizient zeigt, dass die Anzahl der überlebenden
Unternehmen eine sehr hohe relative Streuung aufweist. Die
Überlebensrate hingegen ist vergleichsweise homogen verteilt, was auf
ähnliche Überlebenswahrscheinlichkeiten über die betrachteten
Wirtschaftszweige hinweg hindeutet.


### Interpretation der Lage- und Streuungskennwerte

Die statistischen Kennwerte zeigen deutliche Unterschiede in der
Verteilung der beiden betrachteten Variablen.

Die Anzahl überlebender Unternehmen weist einen arithmetischen
Mittelwert von 65,04 auf, während der Median mit 22,5 deutlich darunter
liegt. Dies deutet auf eine rechtsschiefe Verteilung hin, bei der wenige
Wirtschaftszweige eine sehr hohe Anzahl überlebender Unternehmen
aufweisen, während die Mehrheit deutlich niedrigere Werte besitzt. Der
Modus von 1 bestätigt, dass der kleinste Wert am häufigsten vorkommt.

Die Überlebensrate der Unternehmen zeigt hingegen eine deutlich
homogenere Verteilung. Der arithmetische Mittelwert von 88,94 % und der
Median von 90,3 % liegen nahe beieinander, was auf eine relativ
symmetrische Verteilung hindeutet. Der Modus von 100 % zeigt, dass viele
Wirtschaftszweige eine sehr hohe Überlebensrate aufweisen.

Auch die Streuungsmaße bestätigen diese Beobachtungen. Die hohe
Stichprobenvarianz sowie der Variationskoeffizient von 1,38 bei der
Anzahl überlebender Unternehmen weisen auf eine starke relative
Streuung hin. Die Überlebensrate besitzt hingegen eine deutlich
geringere Varianz und einen niedrigen Variationskoeffizienten von 0,12,
was auf eine vergleichsweise geringe Streuung hindeutet.

Zusammenfassend lässt sich festhalten, dass sich die absolute Anzahl
überlebender Unternehmen stark zwischen den Wirtschaftszweigen

## R3.16 – Curve Fitting

In diesem Abschnitt wird eine Ausgleichsfunktion für die im Scatterplot
dargestellten Daten bestimmt. Ziel ist es, einen möglichen allgemeinen
Trend zwischen der Anzahl überlebender Unternehmen und der
Überlebensrate zu beschreiben.


```python
import numpy as np
import matplotlib.pyplot as plt

# x- und y-Werte
x = df["anzahl_ueberlebende"].values
y = df["ueberlebensrate_prozent"].values

# Lineares Curve Fitting (Grad 1)
koeffizienten = np.polyfit(x, y, deg=1)
a, b = koeffizienten

a, b
```

```python
# Werte für die Ausgleichsgerade
x_fit = np.linspace(x.min(), x.max(), 100)
y_fit = a * x_fit + b

plt.figure()

plt.scatter(
    x,
    y,
    label="Messdaten"
)

plt.plot(
    x_fit,
    y_fit,
    label="Lineare Ausgleichsgerade"
)

plt.xlabel("Anzahl überlebender Unternehmen")
plt.ylabel("Überlebensrate in Prozent")
plt.title("Scatterplot mit linearer Ausgleichsgerade")

plt.legend()  

plt.savefig("../figures/scatterplot_curvefit.png", dpi=300, bbox_inches="tight")
plt.show()

```

Für die Analyse des Zusammenhangs zwischen der Anzahl überlebender
Unternehmen und der Überlebensrate wurde ein lineares Curve Fitting
durchgeführt. Hierbei wurde eine lineare Ausgleichsfunktion der Form
y = ax + b mithilfe der Methode der kleinsten Quadrate bestimmt.

Das lineare Modell beschreibt den allgemeinen Trend der Daten, ohne den
Anspruch, jeden einzelnen Messpunkt exakt zu approximieren. Die
ermittelte Steigung der Ausgleichsgeraden ist gering, was darauf
hindeutet, dass nur ein schwacher linearer Zusammenhang zwischen den
beiden Variablen besteht.

Die deutliche Streuung der Messpunkte um die Ausgleichsgerade zeigt,
dass die Überlebensrate der Unternehmen nicht maßgeblich von der
absoluten Anzahl überlebender Unternehmen abhängt. Das Curve Fitting
dient daher primär der Trendbeschreibung und erlaubt keine Aussage über
kausale Zusammenhänge.


## R3.17 – Mathematische Angabe der Ausgleichsfunktion

Die im Rahmen des Curve Fittings bestimmte Ausgleichsfunktion wird in
diesem Abschnitt mathematisch angegeben.


Die lineare Ausgleichsfunktion besitzt die Form

$$
y = a \cdot x + b
$$

Dabei bezeichnet

- $x$ die Anzahl überlebender Unternehmen,
- $y$ die Überlebensrate der Unternehmen in Prozent,
- $a$ die Steigung der Ausgleichsgeraden und
- $b$ den y-Achsenabschnitt.

Für den vorliegenden Datensatz ergeben sich folgende Parameter der
linearen Ausgleichsfunktion:

$$
a = 0{,}0036, \quad b = 88{,}71
$$

Damit lautet die Ausgleichsfunktion:

$$
y = 0{,}0036 \cdot x + 88{,}71
$$


Die geringe Steigung der Ausgleichsgeraden zeigt, dass sich die
Überlebensrate der Unternehmen nur sehr schwach mit zunehmender Anzahl
überlebender Unternehmen verändert. Der y-Achsenabschnitt gibt die
theoretische Überlebensrate bei einer Unternehmensanzahl von null an
und besitzt keine direkte inhaltliche Bedeutung.
